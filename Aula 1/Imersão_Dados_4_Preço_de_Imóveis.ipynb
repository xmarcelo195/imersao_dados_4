##Desafio 6

```python
# Dividir $ do valor
df[['$','Valor']] = df['Valor'].str.split(' ',n=1,expand=True)
```

```python
#ver se tudo é R$
df['$'].unique()
```

```python
df.head(1)
```

```python
df_valor_compra = df[~df.Valor.str.contains('\n')].copy()
#df_valor_compra = df_valor_compra[~df_valor_compra.Valor.str.contains('Ano')]
#df_valor_compra = df_valor_compra[~df_valor_compra.Valor.str.contains('Dia')]
```

```python
df_valor_compra["Valor"] = df_valor_compra["Valor"].str.replace('.','').astype(int)
```

```python
df_valor_compra.describe().style.format({'Valor': '{:,}'})
```

```python
media_precos_bairro = pd.DataFrame(df_valor_compra.groupby('Bairro').mean()['Valor'].sort_values(ascending=False).astype(int)).head(10).style.format(thousands=".")
media_precos_bairro
```
